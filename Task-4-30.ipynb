Определить дробь, числитель которой меньше знаменателя на 2. Если числитель увеличить на 5, а знаменатель
уменьшить на 5, то получится целое число 2.

```python
from sympy.abc import x
from sympy import Eq
from sympy import solve

sol = solve([Eq((x+5)/((x+2)-5), 2)], [x])
print(sol)
f = sol[x] / (sol[x] + 2)
print("Fraction:", f)
```

```python
from sympy.abc import x,y
from sympy import Eq
from sympy import solve

sol = solve([Eq(x-y,-2), Eq((x + 5) / (y - 5), 2)], [x, y])
print(sol)
f = sol[x] / sol[y]
print("Fraction:", f)
```
